```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig


model_id = "google/gemma-7b"


quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.bfloat16
)



tokenizer = AutoTokenizer.from_pretrained(model_id)


model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=quantization_config,
    device_map="auto"
)


print("Model başarıyla yüklendi!")
print(model.config.model_type)

```

```python

case_context = """
Patient Profile:
- Age: 68-year-old female
- Chief Complaint: "White coating in my mouth that I can wipe off"
- Medical History: Type 2 diabetes mellitus, recent antibiotic use for UTI
- Social History: Denture wearer

Clinical Findings:
- Creamy white plaques on tongue dorsum and buccal mucosa
- Plaques can be wiped away revealing underlying erythema

Evaluation Questions:
1. Primary Diagnosis: What is the most likely diagnosis?
2. Predisposing Factors: List three factors that contributed to this condition.
3. Diagnostic Test: What simple chairside test would confirm your diagnosis?
4. Treatment Plan: Outline appropriate antifungal therapy.
"""


prompt = f"""
You are a third-year dental student answering questions about an oral pathology case.
Based on the following case information, provide a clear and concise answer to each question.

Case Information:
{case_context}

Your Answers:
1. Primary Diagnosis:
2. Predisposing Factors:
3. Diagnostic Test:
4. Treatment Plan:
"""
```

```python

inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

outputs = model.generate(**inputs, max_new_tokens=250)

response_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(response_text)
```
